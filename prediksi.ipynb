```python
import os
import glob
import numpy as np
from ultralytics import YOLO

# Path ke folder yang berisi file gambar
folder_path = r'C:\Users\RND\Downloads\GABUNGAN\train\images'

# Mendapatkan semua file dengan ekstensi .jpg dalam folder
image_files = glob.glob(os.path.join(folder_path, '*.jpg'))

# Membuat folder untuk menyimpan file txt
output_folder = os.path.join(folder_path, 'label_txtbaru')
os.makedirs(output_folder, exist_ok=True)

# Iterasi melalui setiap file dalam folder
for image_path in image_files:
    # Load a model
    model = YOLO('runs/detect/train4/weights/best.pt')  # pretrained YOLOv8n model

    # Run batched inference on a list of images
    results = model([image_path])  # return a list of Results objects

    # Process results list
    for result in results:
        boxes = result.boxes  # Boxes object for bbox outputs

    # Konversi tensor PyTorch ke NumPy array
    numpy_array = boxes.xywhn.cpu().numpy()

    # Konversi tensor PyTorch ke NumPy array
    numpy_arraycls = boxes.cls.cpu().numpy()

    # Mendapatkan nama file tanpa ekstensi dan menggabungkannya dengan ekstensi .txt
    base_filename = os.path.splitext(os.path.basename(image_path))[0]
    txt_filename = f'{base_filename}.txt'

    # Path lengkap ke file txt
    txt_filepath = os.path.join(output_folder, txt_filename)

    # Membuka file untuk menulis (mode 'w' berarti write)
    with open(txt_filepath, 'w') as file:
        # Gabungkan data menjadi satu array NumPy
        data = np.column_stack((numpy_arraycls, numpy_array))

        # Urutkan data berdasarkan kolom pertama (x)
        sorted_data = data[data[:, 1].argsort()]

        # Iterasi melalui hasil yang sudah diurutkan dan menulis ke dalam file
        for row in sorted_data:
            cls = int(row[0])
            x, y, w, h = row[1:]
            line = f"{cls} {x} {y} {w} {h}"
            file.write(line + '\n')  # Menambahkan newline untuk setiap baris
```
